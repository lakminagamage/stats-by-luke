```python
import pandas as pd
import seaborn as sns
import requests
import io
from matplotlib import pyplot as plt
```

```python
url='https://raw.githubusercontent.com/lakminagamage/stats-by-luke/main/titanic.csv'
s=requests.get(url).content
df=pd.read_csv(io.StringIO(s.decode('utf-8')))
df.head()
```

```python
# Frequency bar chart
p1 = sns.countplot(x='Survived', data=df,  alpha=0.8, color = "blue")
p1.set(xlabel="Survival conditon", ylabel = "Count",
      title = "Composition of passengers by survival conditon")
```

```python
# Percentage bar chart
# estimator : Statistical function to estimate within each categorical bin.
p2 = sns.barplot(x="Survived",
                 y="Survived",
                 data=df,
                 estimator=lambda x: len(x) / len(df) * 100,
                 color = "blue"
                )
p2.set(xlabel="Survival conditon", ylabel = "Percentage",
      title = "Composition of passengers by survival conditon")
```

```python
#find frequency of each letter grade
survival_count = pd.crosstab(index=df['Survived'], columns='count')
# Create frequency table for survival condition
survival_count  = df['Survived'].value_counts().rename_axis('Type').reset_index(name='counts')
# Add column for percentage
survival_count['Percentage'] = survival_count.counts / survival_count.counts.sum() *100

# Percentage bar chart
p2 = sns.barplot(x="Survived",
                 y="Survived",
                 data=df,
                 estimator=lambda x: len(x) / len(df) * 100,
                 color = "blue"
                )
p2.set(xlabel="Survival conditon", ylabel = "Percentage (%)",
      title = "Composition of passengers by survival conditon")
```

```python
# Two way frequency tab'
twoway = pd.crosstab(index=df['Pclass'], columns=df['Survived'], margins=False).reset_index()
twoway = twoway.rename(columns={0: 'Died', 1: 'Survived'})
twoway

#s1 = sns.barplot(x = 'Pclass', y = 'Survived', data = twoway, color = 'red')
sns.countplot(x='Pclass', data=df,  alpha=0.8, color = "blue")
#s2 = sns.barplot(x = 'Pclass', y = 'Survived', data = twoway, color = 'red')
#s2.set(ylabel = "Count",
      #title = "Composition of passengers by passengar class and survival conditon (red - died , blue- survived)")
```

```python
# Reshape the twoway DataFrame
df1 = pd.melt(twoway, id_vars="Pclass", var_name="Type", value_name="Count")
df1

p = sns.catplot(x='Pclass', y='Count', hue='Type', data=df1, kind='bar')
p.set(ylabel = "Count",
       title = "Composition of passengers by passengar class and survival conditon")
```
